```python
import sympy as sp
import random as rd
import numpy as np
import matplotlib.pyplot as plt
import os
import math
import time
import RSA as rsa
```

```python
def analyze_dec_enc_with_diff_n_sizes():
    rsa.generate_pq_for_n_bits()
    # read the file
    try:
        os.remove("speed_enc.txt")
        os.remove("speed_dec.txt")
    except OSError:
        pass
    with open("n_e_d.txt", "r") as f:
        lines = f.readlines()
        for line in lines:
            n = int(line.split(" ")[1])
            e = int(line.split(" ")[3])
            bit = int(line.split(" ")[7])
            d_ = int(line.split(" ")[5])
            # encrypt the message
            start1 = time.time()
            c = rsa.encode_encrypt_message("doaa magdy ibrahim", e, n)
            end1 = time.time()
            
            # decrypt the message
            start2 = time.time()
            m = rsa.decode_decrypt_message(c, d_, n)
            end2 = time.time()

            # write the time it took to encrypt and decrypt in a file
            with open("speed_enc.txt", "a") as f:
                f.write("bit: " + str(bit) + " time_enc: " + str(end1 - start1) + "\n")
            with open("speed_dec.txt", "a") as f:
                f.write("bit: " + str(bit) + " time_dec: " + str(end2 - start2) + "\n")
            
            # plot the data
        rsa.plot_to_analyze("speed_enc.txt", "Encryption Speed")
        rsa.plot_to_analyze("speed_dec.txt", "Decryption Speed")

analyze_dec_enc_with_diff_n_sizes()
```
